NumPy Trigonometric Functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sin_45 = np.sin(45)
```

```python
print("Value of sin 45 degree",sin_45)
```

```python
sin_45_radiance = np.sin(45*np.pi/180)
sin_45_radiance
```

```python
cos_45 = np.sin(45)
cos_45
```

```python
cos_45_radiance = np.cos(45*np.pi/180)
cos_45_radiance
```

```python
tan_45 = np.sin(45)
tan_45
```

```python
tan_45_radiance = np.tan(45*np.pi/180)
tan_45_radiance
```

```python
# GRAPHICAL REPRESENTATION OF TRIGNOMETRIC FUNCTION
```

```python
x = np.arange(0,5*np.pi,0.1)
y = np.sin(x)
plt.plot(x,y)
plt.title("SIN GRAPH")
plt.xlabel("X AXIS")
plt.ylabel("Y AXIS")
plt.show()
```

```python
x = np.arange(0,5*np.pi,0.1)
y = np.cos(x)
plt.plot(x,y)
plt.title("COS GRAPH")
plt.xlabel("X AXIS")
plt.ylabel("Y AXIS")
plt.show()
```

```python
x = np.arange(0,10*np.pi,0.1)
y = np.tan(x)
plt.plot(x,y)
plt.title("TAN GRAPH")
plt.xlabel("X AXIS")
plt.ylabel("Y AXIS")
plt.show()
```
