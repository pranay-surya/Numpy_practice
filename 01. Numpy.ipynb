### Creating and Checking No. Element ,Shape and Dimension of Array

```python
import numpy as np
```

#### ONE DIMENTIONAL ARRAY

```python
arr_1 = np.array([1,2,3,4,5,6,7])
arr_1
```

#### TWO DIMENTIONAL ARRAY

```python
arr_2 = np.array([[1,2,3],[1,2,3],[1,2,3]])
arr_2
```

#### THREE DIMENTIONAL ARRAY

```python
 arr_3 = np.array([
     [
         [1,2,3],
         [1,2,3],
         [1,2,3]
     ],
     [
         [1,2,3],
         [1,2,3],
         [1,2,3]
     ]   
 ])
arr_3
```

#### Total Element in Array

```python
print("Size of Array 1 :: ",arr_1.size)  
print("Size of Array 2 :: ",arr_2.size)   
print("Size of Array 3 :: ",arr_3.size)   
```

 #### Shape of Array

```python
print("Shape of Array 1 ::",arr_1.shape)
print("Shape of Array 2 ::",arr_2.shape)
print("Shape of Array 3 ::",arr_3.shape)
```

#### Dimension of Array

```python
print("Dimension of Array 1 ::",arr_1.ndim)
print("Dimension of Array 2 ::",arr_2.ndim)
print("Dimension of Array 3 ::",arr_3.ndim)
```
