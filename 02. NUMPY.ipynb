```python
import numpy as np
```

#### CREATING EMPTY ARRAY

```python
arr1 = np.empty((3,3),dtype = int)   
arr1
```

#### CREATING ARRAY WITH ONES

```python
arr2 = np.ones((3,3),dtype = int)  
arr2
```

#### CREATING ARRAY WITH ZEROS

```python
arr3 = np.zeros((3,3))    
arr3
```

 #### CONVERTING ARRAY INTO BOOLEAN FORM

```python
arr4 = np.ones((3,3),dtype = bool)  
arr4
```

#### CONERTING ARRAY INTO STRING FORM

```python
arr5 =  np.ones((3,3),dtype = str)   
arr5
```
