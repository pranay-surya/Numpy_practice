### NUMPY ARRAY SLICING 

```python
import numpy as np
```

```python
arr1 = np.arange(1,101).reshape(10,10)
arr1
```

#### For Accessing The specific Element in array

```python
 arr1[0,9]
```

#### Slicing in Array 

```python
arr1[:,1:5] # It Select All Rows and Selected cols.
```

```python
arr1[1:4,:] # It select all cols and selected rows
```

#### Negative Slicing 

```python
arr1[:]
```

```python
arr_ne = arr1[-6:-1,-5:-1]
arr_ne
```

```python
arr_ne.ndim
```

```python
arr1[:,0]
```

```python
arr1[:,0:1]
```
