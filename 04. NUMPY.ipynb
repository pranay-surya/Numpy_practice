### Numpy Mathematical Functions

```python
import numpy as np
```

#### APPLY MATHS OPERATION ON ARRAY

```python
arr1 = np.arange(1,10).reshape(3,3)
arr2 = np.arange(11,20).reshape(3,3)
```

#### Addition

```python
arr_add = np.add(arr1,arr2) 
arr_add
```

#### Subtraction

```python
arr_sub = np.subtract(arr1,arr2)
arr_sub
```

#### Multiplication

```python
arr_mul = np.multiply(arr1,arr2,arr1)
arr_mul
```

#### Division

```python
arr_div = np.divide(arr1,arr2)
arr_div
```

#### Modulo

```python
arr_mod = np.mod(arr1,arr2)
arr_mod
```

```python
arr_dot = np.dot(arr1,arr2)   
arr_dot
```

```python
print("Sum of Array ::",np.sum(arr1))
print("Max Element in Array ::",np.max(arr1))
print("Max Element in Array ::",np.max(arr1))
print("index of max Element ::",np.argmax(arr1))
print("index of min Element ::",np.argmin(arr1))
print("Standart deviation of Array ::",np.std(arr1))
print("Mean of Array ::",np.mean(arr1))
```

```python
np.sqrt(arr1)
```
