# Python NumPy String Operations Methods

```python
import numpy as np
```

```python
str1 = "Local"
str2 = "host"
str3 = "DUMP"
arr1 = np.char.add(str1,str2) # add two string in array  
arr1
```

```python
type("arr1")
```

```python
arr2 = np.char.multiply(str1,5) # Repeat the string multiple times
arr2
```

```python
arr3 = np.char.capitalize(str2) # Convert 1'st letter into capital case
arr3
```

```python
arr4 = np.char.lower(str3) # Lower case the all letter
arr4
```

```python
arr5 = np.char.upper(str2) # Convert all letters to upper case
arr5
```

```python
arr6 = np.char.title(arr1)  # convert string into title form
arr6
```

```python
arr7 = np.char.center(str1,20,fillchar="-")  #function set the ng in the center of given string length. If you want to fill the ngâ€™s side black space. Then give fillchar parameter with character.
arr7
```

```python
arr8 = np.char.split(arr1,sep= "l")
arr8
```

```python
arr9 = np.char.splitlines(arr1,"/")
arr9
```

```python
arr10 = np.char.join([":","/"],["dmy","dmy"])
arr10
```

```python
arr11 = np.char.replace(arr1,"host","lock")
arr11
```
