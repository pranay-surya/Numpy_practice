### NumPy Array Concatenation and Split 

```python
import numpy as np
```

#### Numpy arange() and reshape()

```python
arr1 = np.arange(1,21).reshape(4,5)
arr1
```

```python
arr2 = np.arange(1,21).reshape(4,5)
arr2
```

#### join two array

```python
arr3 = np. concatenate((arr1,arr2))
arr3
```

```python
arr3 = np. concatenate((arr1,arr2),axis= 1) # AXIS , 1 = ROW
arr3
```

#### JOIN TWO ARRAY BY STACK

```python
np.vstack((arr1,arr2))
```

```python
np.hstack((arr1,arr2))
```

#### SPLIT ARRAY

```python
split_arr1 = np.split(arr1,2)
split_arr1 
```

#### ACCESS THE SPLIT ARRAY

```python
split_arr1[0]
```

```python
sp_split_arr1 = np.split(split_arr1[0],2)
sp_split_arr1
```

```python
sp_split_arr1[0]
```

#### SPLIT 1 DIMENTIONAL ARRAY

```python
arr_1D = np.arange(10)
arr_1D
```

```python
np.split(arr_1D,(2))
```
